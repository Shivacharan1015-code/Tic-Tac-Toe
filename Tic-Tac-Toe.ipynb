```python
#TicTacToe
row1=[7,8,9]
row2='---------'
row3=[4,5,6]
row4='---------'
row5=[1,2,3]
#display
def display():
    print(row1)
    print(row2)
    print(row3)
    print(row4)
    print(row5)
    

#userinput
u=0
l=['1','2','3','4','5','6','7','8','9']
def userinput():
    x='a'
    global u
    global l
    while x.isdigit()==False:
        if u%2==0:
            print("\nplayer 1, it's your turn")
        else:
            print("\nplayer 2, it's your turn")
        x=(input('Enter your choice (1-9)  '))
        
        if x not in l:
            print('Out of range')
            x='a'
        
        if x.isdigit()==True:
            u=u+1
            l.remove(x)
            return int(x)
```

```python
#change in display
def changedisplay(x):
    global u
    x=str(x)
    if x=='7':
        if u%2==0:
            row1[0]='O'
        else:
            row1[0]='X'
    
    if x=='8':
        if u%2==0:
            row1[1]='O'
        else:
            row1[1]='X'
    
    if x=='9':
        if u%2==0:
            row1[2]='O'
        else:
            row1[2]='X'
            
    if x=='4':
        if u%2==0:
            row3[0]='O'
        else:
            row3[0]='X'
            
    if x=='5':
        if u%2==0:
            row3[1]='O'
        else:
            row3[1]='X'
    
    if x=='6':
        if u%2==0:
            row3[2]='O'
        else:
            row3[2]='X'
            
    if x=='1':
        if u%2==0:
            row5[0]='O'
        else:
            row5[0]='X'
            
    if x=='2':
        if u%2==0:
            row5[1]='O'
        else:
            row5[1]='X'
            
    if x=='3':
        if u%2==0:
            row5[2]='O'
        else:
            row5[2]='X'
            
    print("------------------------------------------------------")
```

```python
#row check
def T_check():        
    if row1[0]==row1[1] and row1[1]==row1[2]:
        return True
    elif row1[0]==row3[0] and row3[0]==row5[0]:
        return True
    else:
        return False
        
def L_check():
    if row1[2]==row3[2] and row3[2]==row5[2]:
        return True
    elif row5[0]==row5[1] and row5[1]==row5[2]:
        return True
    else:
        return False
    
def HAV_check():
    if row1[1]==row3[1] and row3[1]==row5[1]:
        return True
    elif row3[1]==row3[0] and row3[0]==row3[2]:
        return True
    else:
        return False
    
def X_check():
    if row3[1]==row1[0] and row1[0]==row5[2]:
        return True
    elif row3[1]==row1[2] and row1[2]==row5[0]:
        return True
    else:
        return False
```

```python
#check winner
def checkwinner(o):
    global u
    if u>=3:
        x=str(o)
        if x=='7':
            if T_check() or X_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                    
        elif x=='8':
            if HAV_check() or T_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
        elif x=='9':
            if L_check() or X_check() or T_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                    
        elif x=='4':
            if T_check() or HAV_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                    
        elif x=='5':
            if HAV_check() or X_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                    
        elif x=='6':
            if L_check() or HAV_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                    
        elif x=='1':
            if T_check() or X_check() or L_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                    
        elif x=='2':
            if L_check() or HAV_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                    
        elif x=='3':
            if L_check() or X_check():
                if u%2==0:
                    print('Player 2 has WON!!!')
                    return True
                else:
                    print('Player 1 has WON!!!')
                    return True
                
        else:
            return False
    else:
        return False
```

```python
stop=False
c=0
display()
while not stop:
    c=c+1
    o=userinput()    
    changedisplay(o)
    display()
    stop=checkwinner(o)    
    if not stop and c==9:
        print("It's a Tie")
        stop=True
```
